```python
import pandas as pd 
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as px 

df_asian_country = pd.read_csv('Countries-Asia.csv')
```

```python
continent = ['Asia', 'Europe', 'North America', 'South America', 'Africa', 'Antarctica', 'Australia']

df = pd.read_csv('population.csv')
df.rename(columns = {'Total population (Gapminder, HYDE & UN)':'Population'}, inplace=True)

df_sorted = df[df['Year'] >= 1900].drop(columns='Code').sort_values('Population', ascending=False)
df_continent = df_sorted[df_sorted['Entity'].isin(continent)].rename(columns = {'Entity':'Continent'})
```

```python
fig_continent = px.line(df_continent, x='Year', y='Population', color='Continent', title='World Population by Continent From 1900 to 2021')
fig_continent.show()
```

Analyzing each country in Asia

```python
df_asia = df[df['Entity'].isin(df_asian_country['name'])].sort_values('Population', ascending=False)
df_asia = df_asia[df_asia['Year'] == 2021]
```

```python
fig_asia = px.bar(df_asia.rename(columns = {'Entity':'Country'}), x='Country', y='Population', color='Country', title='Population in Asia (2021)')
fig_asia.show()
```

Which country is the most populated in Asia?

```python
total_asia = df_continent[df_continent['Continent'] == 'Asia'].Population
df_asia['Percentage'] = df_asia['Population']/df_asia['Population'].sum()*100

df_asia_sorted = df_asia[:12].append({'Entity':'Other (smaller than 1%)', 'Percentage':100-df_asia['Percentage'][:12].sum()}, ignore_index=True)

fig_asia_percentage = px.pie(df_asia_sorted.rename(columns={'Entity':'Country'}).sort_values('Percentage', ascending=False), names='Country', values='Percentage', color='Country', title='Percentage Population of Countries in Asia (2021)')
fig_asia_percentage.show()
```

Even though Asia has the largest population in the world, the majority of population are from China and India. This can be important later on when we discuss about population density and over-population crisis.

Population Density graph

```python
import pandas as pd
import plotly.express as px 

# Read DataFrame
density_df=pd.read_csv('population-density.csv')
```

```python
density_df.Density = pd.to_numeric(density_df.Density)                                                    # Convert to numeric
density_df = density_df.rename(columns={'Population density (people per sq. km of land area)':'Density'}) # Rename the column
density_df = density_df.groupby(['Entity','Year']).mean()['Density'].reset_index()                        # Groupby Entity and Year
```

```python
# Show the graph
fig = px.line(density_df, x='Year', y='Density',color='Entity',title = 'Population density (people per sq. km of land area) From 1961 to 2018')
fig.show()  
```

## Population Density in Thailand and the world

```python
# Filter countries
clist = ['Thailand','World']
densfiltered_df = density_df[density_df.Entity.isin(clist)]
densfiltered_df = densfiltered_df.groupby(['Entity','Year']).mean()['Density'].reset_index()
```

```python
#Show the graph
fig = px.line(densfiltered_df, x='Year', y='Density', title='Population Density in Thailand and the world',color='Entity')
fig.show()
```

## Top 10 most densely populated countries in 2018

```python
df11 = density_df[density_df['Year']==2018]                                     # Create DataFrame by selecting only 2018
df11= df11.sort_values(['Density'],ascending=False).reset_index()               # Sort values by Density
df11 = df11[:10]                                                                # Select Top ten countries
```

```python
# Show the graph
fig = px.bar(df11, x='Entity', y='Density',title ='Top 10 most densely populated countries in 2018',color='Density')
fig.show()
```

Many of the world’s small island or isolated states have large populations for their size. Macao, Monaco, Singapore, Hong Kong and Gibraltar are the five most densely populated. Singapore has nearly 8,000 people per km2 – more than 60 times as dense as Thailand.

# 3. Is the rate of population growth increasing or decreasing?

Plain Population Graph

```python
import pandas as pd
```

```python
! wget https://datahub.io/core/country-list/r/data.csv
! gdown --id 13c6rhqEDvVacat7HwssBLGsQsX8QGurM        #source https://ourworldindata.org/grapher/population
```

```python
#get list of countries
country_df=pd.read_csv('data.csv')
country_list=[country for country in country_df['Name']]                        #from df to list
```

```python
#create df and filter for countries in 1800+
df0=pd.read_csv('population.csv')                                               
df0=df0.rename(columns={'Total population (Gapminder, HYDE & UN)':'Population'})#rename column for easier access
df0=df0[df0['Year'] >= 1800]                                                    #remove data of years less than 1800
df=df0[df0['Entity'].isin(country_list)]                                        #only included country in country list 
```

## Rough Total population of each country throughout 1800-2020

```python
import plotly.express as px

fig=px.line(df,x='Year',y='Population',color='Entity', title='Total population of each country throughout 1800-2020')
fig.show()
```

```python
#df1 = pivot year 1949+ of df
df1=df0[df0['Year'] >= 1949].pivot(index='Year',columns='Entity',values='Population')                 #set pivot
df1=df1.dropna(axis=1)                                                          #remove countries with incomplete data
```

```python
pctc_df=df1.pct_change()                                                        #find percent change of values
pctc_df=pctc_df.multiply(100)                                                   #times 100
pctc_df=pctc_df.dropna(axis=0)                                                  
pctc_df['Year']=pctc_df.index                                                   #since 'Year' was used as an index, when pivoting df1, 'Year' is gone as an column. Creating new one for the graph
pctc_df=pctc_df.melt(id_vars=['Year'], value_name='Percent Change')             #melt the table. found it quite hard to understand, but it did the job https://pandas.pydata.org/docs/reference/api/pandas.melt.html
pctc_df['Entity']=[str(i) for i in pctc_df['Entity']]
```

## Population increase/decrease rate per year for selected population

```python
from IPython.display import clear_output
import time

regions=['Thailand', 'Myanmar', 'Philippines', 'Malaysia', 'Singapore']         #default selection
pctc_df_filtered=pctc_df[pctc_df['Entity'].isin(regions)]                       #filter
while True:             
  fig1=px.line(pctc_df_filtered,
               x='Year',
               y='Percent Change',
               color='Entity',
               title='Population increase/decrease rate per year for selected countries',
               #color_discrete_sequence=px.colors.qualitative.Light24            #https://plotly.com/python/discrete-color/#color-sequences-in-plotly-express
               )
  fig1.show()                                                                   
  
  print(
      'Countries and Continents selection. Example: ',
      '\n\tThailand, Myanmar, Philippines, Malaysia, Singapore',
      '\n\tThailand, United States, China, Japan, India',
      '\n\tWorld, Asia, Europe, South America, North America, Africa, Antarctica, Australia',
      '\n\tThailand, World, Asia'
      '\n\'High\' and \'Low\' for top 5 highest/lowest population growth rate in 2021'
      '\n\'End\' to quit'
      )
  time.sleep(0.7)                                                               #without time delay the input box does not show up, probably due to it being created before fig1 shows
  
  input_str=input()
  if input_str == 'End' or input_str == '':                                     #end the iteration 
    break;        
                                                                                #prepare string to list                                                               
  regions=input_str.split(',')                                                
  regions=[word.strip() for word in regions]                                  
  clear_output()
                                                                                #check if contain special words
  haveHigh='High' in regions
  haveLow='Low' in regions
  if haveHigh or haveLow:
    top_list=[]
    pctc_df_grouped=pctc_df[pctc_df['Year'] == 2021].groupby('Percent Change')
    top_list_initial=[i for i in pctc_df_grouped.first()['Entity']]
    if haveHigh:
      top_list=top_list_initial[-5:]                                            #High and Low are inverse because the list are ranked least to highest
    if haveLow:
      top_list=top_list+top_list_initial[0:5]
    regions=regions+top_list
                                                                                #add regions
  pctc_df_filtered=pctc_df[pctc_df['Entity'].isin(regions)]                  
```
